# Data Preparation - Superstore

This notebook prepares the superstore dataset for further analysis.

#### The following steps are performed:
- loading source csv files
- converting date columns to proper datetime format
- merging tables into a sales_final table
- checking accuracy of the sales_final table

```python
import pandas as pd
```

```python
customers = pd.read_csv("../data/dim_customer.csv")
products = pd.read_csv("../data/dim_product.csv")
sales = pd.read_csv("../data/fact_sales.csv")
```

### Data Transformation

Below section of this notebook:
- converts date columns into the appropriate datetime format to enable time-based analysis
- joins the facts table with dimension tables using a left join
- counts rows of the initial and final sales table to ensure no data was duplicated or lost

This is to ensure that the uploaded data is reliable and appropriate for further transformation and analysis.

```python
sales["order_date_day"] = pd.to_datetime(sales["order_date_day"])
```

```python
sales["ship_date_day"] = pd.to_datetime(sales["ship_date_day"])
```

```python
sales_customers = pd.merge(sales, customers, on='customer_id', how='left')
```

```python
sales_final = pd.merge(sales_customers, products, on='product_id', how='left')
```
